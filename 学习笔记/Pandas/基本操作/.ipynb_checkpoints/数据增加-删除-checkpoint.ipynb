```python
import pandas as pd
data=[[45,65,100],[56,45,50],[67,67,67]]
index=['zhang','li','wang']
columns=['数学','语文','英语']
df=pd.DataFrame(data=data,index=index,columns=columns)
```

```python
df
```

```python
#列直接插入
df['政治']=[88,72,92]
df
```

```python
df.insert(1,'历史',[92,82,79])#指定列的位置插入数据
df
```

```python
df.loc['zhu']=[18,28,39,77,92]#行增加
```

```python
df
```

```python
#修改列标题
df.columns=['语言','数学','政治','英语','历史']
df
```
